**Advanced Housing Prices- Feature Engineering**

The main aim of this project is to predict the house price based on various features which we will discuss as we go ahead



**We will be performing all the below steps in Feature Engineering**

1. Missing values
2. Temporal variables
3. Categorical variables: remove rare labels
4. Standarise the values of the variables to the same range


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dataset=pd.read_csv('train.csv')
```

checking for missing values in Categorial feature

```
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().values.any() and dataset[feature].dtype=='O']
for feature in feature_nan:
    print(feature,'  ',dataset[feature].isnull().sum(),'Missing')
```

**Hence we have Missing values in Categorial features we need to handle it**

**we will replace all the Nan values with *'Missing'* label**




```
def replace_cat_feature(feature_nan,dataset):
  
  for feature in feature_nan:
    dataset[feature]=dataset[feature].fillna('Missing')
  return dataset

dataset=replace_cat_feature(feature_nan,dataset)
dataset[feature_nan].isnull().sum()
```

```
dataset[feature_nan].head(10)
```

Filling missing value in numerical feature

```
feature_num_nan=[feature for feature in dataset.columns if dataset[feature].isnull().values.any() and dataset[feature].dtype !='O']
for feature in feature_num_nan:
  print(feature,' ',dataset[feature].isnull().values.sum())
```

```
def replace_num_feature(feature_num_nan,dataset): 
  for feature in feature_num_nan:
    median_value=dataset[feature].median()
    dataset[feature]=dataset[feature_num_nan].fillna(median_value)
  return dataset

dataset=replace_num_feature(feature_num_nan,dataset)
dataset[feature_num_nan].isnull().sum()
```

```
dataset.head(50)
```

```
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
  dataset[feature]=dataset['YrSold']-dataset[feature]
```

```
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head(10)
```

```
# checking if there is any Nan values in dataset
dataset.isnull().values.any()
```

**Now in our dataset train.csv we do not have any missing values**

**Numerical variable**

Since the numerical variables are skewed we will perform log normal distribution

```
for feature in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']:
  dataset[feature]=np.log(dataset[feature])
```

**Handling Rare Categorial variable**

We will remove categorical variables that are present less than 1% of the observations

```
categorial_feature=[feature for feature in dataset.columns if dataset[feature].dtype =='O']
```

```
for feature in categorial_feature:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```
dataset.head()
```

```
for feature in categorial_feature:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```
dataset.head()
```

```
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
scaling_feature
```

```
from sklearn.preprocessing import MinMaxScaler
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
scaler=MinMaxScaler()
scaler.fit_transform(dataset[feature_scale])
```

```
data=pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),pd.DataFrame(scaler.fit_transform(dataset[feature_scale]),columns=feature_scale)],axis=1)
```
